```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(2, 2, figsize = (20, 5))
axes[0, 0].plot(x, y, label = "$sin(x)$")
axes[0, 0].title.set_text('Relacion X - Y')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')
axes[0, 0].legend()

axes[0, 1].plot(y, x, label = "$sin(y)$")
axes[0, 1].title.set_text('Relacion Y - X')
axes[0, 1].set_xlabel('Y')
axes[0, 1].set_ylabel('X')
axes[0, 1].plot(y, x)
axes[0, 1].legend()
```

```python
# Se puede personalizar la posición de la leyenda
fig, ax1 = plt.subplots(1)
ax1.plot(x,y, 'b', label = "sin(x)")
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
```
