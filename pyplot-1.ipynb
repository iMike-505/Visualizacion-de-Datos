```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
plt.plot(x, y, 'r:')
plt.show()
```

```python
# Crear histograma
plt.hist(y)
plt.show()
```

```python
# Crear grafica de pie
plt.pie(x)
plt.show()
```

```python
# Correlacion entre variables
plt.scatter(x, y)
plt.show()
```

```python
# Distribucion de datos: Grafico de caja
plt.boxplot(x)
plt.show()
```
