```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Hacemos un array de n√∫meros aleatorios
data = np.random.randint(1, 50, 100)
```

# Histograma

```python
plt.hist(data,bins=10,histtype='bar')
plt.show()
```

```python
# Bins es el n√∫mero de barras
plt.hist(data, bins = 50, histtype = 'step')
plt.show()
# histtype cambia el estilo del histograma
```

# Boxplot

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
plt.show()
```

# Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area =(30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
# area = tamanio
# alpha para la transparencia
plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```
