```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r--')
plt.plot(y, x, 'b:')
plt.subplot(2, 1, 2)
plt.pie(y)
plt.show()
```

# Orientado a objetos

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # Los primeros numeros hacen referencia a la posicion de la figura

axes.plot(x, y, 'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.17,0.55,0.4,0.3])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
plt.show()
```
