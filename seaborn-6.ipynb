```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Joinplot
Joinplot une 2 diferentes gr√°ficas en una sola

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist', marginal_ticks=True, #muestra una peque√±a tabla para la gr√°fica exterior
marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge')); #marginal_kws se le pueden agregar diferentes par√°metros
```

# Pairplot
Pairplot correlaciona todas as vriables num√©ricas que hay en el dataset

```python
sns.pairplot(data = tips);
```

Solo se muestra una esquina del pairplot, eliminando as√≠ las gr√°ficas repetidas

```python
sns.pairplot(data = tips, corner = True)
```
