# Proyección Estereográfica

Un macizo rocoso está afectado por tres
familias de discontinuidades: F1 40°/30°, F2 65°/150° y E
20°/200°. 
Obtener las cuñas posibles para un talud con
orientación 76°/105°.

```python
import matplotlib.pyplot as plt
import mplstereonet

def plot_stereonet(data, title):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='stereonet')

    strikes, dips = zip(*data)
    ax.plane(strikes, dips, c='b', label='Datos')

    ax.grid(True)
    ax.legend()
    ax.set_title(title)

    plt.show()

# Datos de las familias de discontinuidades
f1_strike = 40
f1_dip = 30

f2_strike = 65
f2_dip = 150

e_strike = 20
e_dip = 200

# Datos del talud
talud_strike = 76
talud_dip = 105

# Convertir los datos a coordenadas esféricas
data = [
    (f1_strike, f1_dip),
    (f2_strike, f2_dip),
    (e_strike, e_dip),
    (talud_strike, talud_dip)
]

# Realizar la proyección estereográfica
plot_stereonet(data, "Proyección estereográfica")
```
