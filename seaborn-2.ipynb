```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot(data = tips, x = 'total_bill')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
# hue hace un desgloce
sns.displot(data = tips, x = 'total_bill', y = 'tip', hue ='sex')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', hue ='sex')
plt.show()
```

```python
# Kind define con que tipo de grafica trabajar
sns.displot(data = tips, x = 'total_bill', hue ='sex', kind = 'kde', legend = True, palette = 'dark', alpha = .2)
plt.show()
```
